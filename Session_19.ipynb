Question 05

```python
def histplot():
  plt.subplot(2, 2, 1)
  sns.histplot(df['price'])
  plt.title('Price')

  plt.subplot(2, 2, 2)
  sns.histplot(df['area'])
  plt.title('Area')

  plt.subplot(2, 2, 3)
  sns.histplot(df['bedrooms'])
  plt.title('Bedrooms')

  plt.subplot(2, 2, 4)
  sns.histplot(df['stories'])
  plt.title('Stories')

  plt.tight_layout()

histplot()
```

Insights

1. Price: Most of the house have price below 5M.
2. Area: Many houses have plot sizes around 4000–7000 sq.ft.
3. Bedrooms: The majority of houses have 3 bedrooms, followed by 2 and 4.
4. Stories: Most houses are 1–2 stories high. A smaller group has 3–4 stories.

Question 06

```python
def countplot():
  plt.subplot(2, 2, 1)
  sns.countplot(data=df, x='mainroad')
  plt.title('Main Road')

  plt.subplot(2, 2, 2)
  sns.countplot(data=df, x='guestroom')
  plt.title('Guestroom')

  plt.subplot(2, 2, 3)
  sns.countplot(data=df, x='prefarea')
  plt.title('Prefered Area')

  plt.subplot(2, 2, 4)
  sns.countplot(data=df, x='furnishingstatus')
  plt.xticks(rotation=45)
  plt.title('Furnishing Status')

  plt.tight_layout()

countplot()
```

Insights

1. Main Road: Most houses are on the main road. Easy access makes them more attractive.
2. Guestroom: Few houses have a guestroom. Rare feature, can be sold as premium.
3. Preferred Area: More houses are outside preferred areas. Houses in preferred areas are fewer, so they can be priced higher.
4. Furnishing Status: Semi‑furnished houses are most common. Furnished houses are fewer and can be marketed as ready‑to‑move.
